#Theory Questions

##1. What is Hypothesis Testing in Statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on a sample of data.

### ‚úÖ Key Terms:
- **Null Hypothesis (H‚ÇÄ):** A general statement or default position that there is no effect or no difference.
- **Alternative Hypothesis (H‚ÇÅ or Ha):** A statement that contradicts the null hypothesis.

### üéØ Purpose:
To determine whether the observed data provides enough evidence to reject the null hypothesis.

### üìä Common Tests:
- **Z-test** (when population variance is known)
- **T-test** (small sample sizes, unknown variance)
- **Chi-square test** (for categorical data)
- **ANOVA** (compare more than two means)

### üîç Steps in Hypothesis Testing:
1. Formulate H‚ÇÄ and H‚ÇÅ
2. Choose significance level (Œ±), typically 0.05
3. Select the appropriate test
4. Compute test statistic
5. Compare with critical value or p-value
6. Make a decision (Reject or Fail to Reject H‚ÇÄ)

### üìå Example:
Testing whether a sample mean is significantly different from a known population mean.

---
##2.  What is the null hypothesis, and how does it differ from the alternative hypothesis ?

In hypothesis testing, we use two competing statements to evaluate a claim:

### üìå Null Hypothesis (H‚ÇÄ):
- A statement of **no effect**, **no difference**, or **status quo**.
- It assumes that any kind of difference or significance you see in your data is due to chance.
- We **test** this assumption with the data.

**Example:**  
H‚ÇÄ: The average test score of students is 75.

---

### ‚ö° Alternative Hypothesis (H‚ÇÅ or Ha):
- A statement that **contradicts the null hypothesis**.
- It suggests that there **is** an effect or a difference.
- If evidence from the data is strong enough, we **reject H‚ÇÄ in favor of H‚ÇÅ**.

**Example:**  
H‚ÇÅ: The average test score of students is **not** 75.

---

### üéØ Summary Table:

| Feature                 | Null Hypothesis (H‚ÇÄ)       | Alternative Hypothesis (H‚ÇÅ)  |
|------------------------|----------------------------|------------------------------|
| Assumes                | No change or effect        | There is a change or effect  |
| Decision Based On      | p-value & significance     | Rejection of H‚ÇÄ              |
| Goal                   | To test its validity       | To find evidence against H‚ÇÄ  |

---
##3.  What is the significance level in hypothesis testing, and why is it important ?


### üìå Definition:
The **significance level**, denoted by **Œ± (alpha)**, is the **probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true**.

In simpler terms, it's the **threshold for how much risk we're willing to take** of making a **Type I error** (false positive).

---

### ‚úÖ Common Significance Levels:
- **Œ± = 0.05** ‚Üí 5% risk of wrongly rejecting H‚ÇÄ
- **Œ± = 0.01** ‚Üí 1% risk (more strict)
- **Œ± = 0.10** ‚Üí 10% risk (less strict)

---

### üß™ Why is it Important?

- Sets the **critical boundary** for decision-making.
- Helps control the risk of **incorrect conclusions**.
- Balances **sensitivity** (detecting real effects) with **specificity** (avoiding false positives).

---

### üìä Example:
If p-value < Œ± ‚Üí **Reject H‚ÇÄ**  
If p-value ‚â• Œ± ‚Üí **Fail to reject H‚ÇÄ**

Choosing a lower Œ± makes the test **more conservative** (harder to reject H‚ÇÄ), while a higher Œ± increases the **chance of finding significance** but also the risk of error.

---
##4.  What does a P-value represent in hypothesis testing ?

## üîç What Does a P-value Represent in Hypothesis Testing?

### üìå Definition:
The **P-value** (probability value) is the **likelihood of observing the sample data, or something more extreme, assuming the null hypothesis (H‚ÇÄ) is true**.

---

### üéØ Interpretation:

- A **small p-value** (typically ‚â§ 0.05) suggests **strong evidence against** the null hypothesis.
- A **large p-value** (> 0.05) suggests **weak evidence** against the null hypothesis ‚Üí we **fail to reject** it.

---

### üö¶ Decision Rule:

| P-value           | Conclusion                        |
|------------------|------------------------------------|
| p ‚â§ Œ± (e.g., 0.05) | Reject the null hypothesis (H‚ÇÄ)   |
| p > Œ±            | Fail to reject the null hypothesis |

---

### üìå Example:
If a test returns a **p-value of 0.03**, and you‚Äôre using a **significance level of 0.05**, you reject the null hypothesis because 0.03 < 0.05.

This means the observed data would be **very unlikely** under the assumption that H‚ÇÄ is true.

---
##5.  How do you interpret the P-value in hypothesis testing ?


### üß™ P-value Interpretation:

The **p-value** tells us the probability of obtaining a result **as extreme as the observed one**, assuming the **null hypothesis (H‚ÇÄ)** is true.

---

### üö¶ Interpretation Rules:

| P-value           | Interpretation                                      | Action                      |
|------------------|-----------------------------------------------------|-----------------------------|
| **p ‚â§ 0.01**      | Very strong evidence against H‚ÇÄ                    | Reject H‚ÇÄ                  |
| **0.01 < p ‚â§ 0.05**| Moderate evidence against H‚ÇÄ                      | Reject H‚ÇÄ                  |
| **0.05 < p ‚â§ 0.10**| Weak evidence against H‚ÇÄ                          | Often fail to reject H‚ÇÄ    |
| **p > 0.10**      | No evidence against H‚ÇÄ                            | Fail to reject H‚ÇÄ          |

---

### ‚úÖ Example Statement:
- If **p = 0.03**, and **Œ± = 0.05**, we say:
  > "Since the p-value is less than 0.05, we reject the null hypothesis. There is statistically significant evidence to support the alternative hypothesis."

---

### ‚ö†Ô∏è Important Note:
A **low p-value does not prove the alternative hypothesis is true**‚Äîit only suggests that the observed data is unlikely under the null hypothesis.

##6.  What are Type 1 and Type 2 errors in hypothesis testing ?

In hypothesis testing, we make a decision to **reject** or **fail to reject** the null hypothesis (H‚ÇÄ).
Sometimes, our decision can be wrong. These wrong decisions are called **errors**.

---

### ‚ùó Type I Error (False Positive)
- Occurs when we **reject the null hypothesis** even though it's **actually true**.
- It's a **false alarm**.
- The probability of making a Type I error is the **significance level (Œ±)**, usually 0.05.

üß† Think: We said something is significant, but it's not.

---

### ‚ùì Type II Error (False Negative)
- Occurs when we **fail to reject the null hypothesis** even though it's **actually false**.
- We **miss a real effect**.
- The probability of making a Type II error is denoted by **Œ≤**.
- Power of a test = **1 - Œ≤** (chance of correctly rejecting H‚ÇÄ)

üß† Think: We missed something that was actually there.

---

### üìä Error Decision Table:

|                    | **H‚ÇÄ True**           | **H‚ÇÄ False**           |
|--------------------|-----------------------|------------------------|
| **Reject H‚ÇÄ**      | ‚ùó Type I Error        | ‚úÖ Correct Decision     |
| **Fail to Reject H‚ÇÄ** | ‚úÖ Correct Decision     | ‚ùì Type II Error        |

---

### üéØ Goal:
- Minimize **both** errors, but reducing one often increases the other.
- Select a **significance level (Œ±)** and ensure **enough sample size** to reduce both risks.

##7. ## üéØ What is the Difference Between a One-Tailed and a Two-Tailed Test in Hypothesis Testing?

In hypothesis testing, the **direction** of the test depends on the **research question** and how the **alternative hypothesis (H‚ÇÅ)** is formulated.

---

### ‚û§ One-Tailed Test (Directional)
- Tests for a change in **one direction only** (either greater than or less than).
- More **powerful** in detecting an effect in that one direction.
- Used when we have a specific directional hypothesis.

**Example:**
- H‚ÇÄ: Œº = 75
- H‚ÇÅ: Œº > 75 ‚Üí Right-tailed test
- H‚ÇÅ: Œº < 75 ‚Üí Left-tailed test

---

### ‚û§ Two-Tailed Test (Non-Directional)
- Tests for any **significant difference**, regardless of direction.
- Splits the significance level (Œ±) between **both tails** of the distribution.
- Used when we are only interested in **any difference**, not direction.

**Example:**
- H‚ÇÄ: Œº = 75
- H‚ÇÅ: Œº ‚â† 75 ‚Üí Two-tailed test

---

### üß† Summary Table:

| Test Type       | Alternative Hypothesis | Critical Region        | Use When                      |
|-----------------|------------------------|------------------------|-------------------------------|
| One-Tailed      | H‚ÇÅ: Œº > Œº‚ÇÄ or Œº < Œº‚ÇÄ   | One side of distribution | Direction of effect matters   |
| Two-Tailed      | H‚ÇÅ: Œº ‚â† Œº‚ÇÄ             | Both sides              | Any difference matters        |

##8. ## üßÆ What is the Z-test, and When is it Used in Hypothesis Testing?

### üìå What is a Z-test?

A **Z-test** is a statistical test used to determine whether there is a significant difference between the **sample mean** and the **population mean**, or between two sample means, **when the population standard deviation is known**.

---

### üìä Formula for One-Sample Z-test:

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:
- \( \bar{x} \) = Sample mean  
- \( \mu \) = Population mean  
- \( \sigma \) = Population standard deviation  
- \( n \) = Sample size  

---

### ‚úÖ When to Use a Z-test?

| Condition                          | Use Z-test? |
|-----------------------------------|-------------|
| Population standard deviation known | ‚úÖ Yes       |
| Large sample size (n ‚â• 30)          | ‚úÖ Preferred |
| Normal distribution assumed         | ‚úÖ Yes       |

---

### üîç Types of Z-tests:
- **One-sample Z-test**: Compare sample mean to population mean.
- **Two-sample Z-test**: Compare means from two independent samples.
- **Z-test for proportions**: Compare proportions (e.g., % support in elections).

---

### üìå Note:
If the population standard deviation is **unknown**, or the sample size is **small (n < 30)**, use a **T-test** instead.

---

##9. ## üìè How Do You Calculate the Z-score, and What Does It Represent in Hypothesis Testing?

### üìå What is a Z-score?

A **Z-score** (or standard score) tells you **how many standard deviations** a data point or sample mean is from the population mean.

---

### üßÆ Formula for Z-score (One-Sample Z-test):

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:
- \( \bar{x} \): Sample mean  
- \( \mu \): Population mean  
- \( \sigma \): Population standard deviation (known)  
- \( n \): Sample size  

---

### üéØ What Does the Z-score Represent?

- The **distance** (in standard deviations) between the sample statistic and the population parameter.
- A **high absolute Z-score** (e.g., > 1.96 for Œ± = 0.05 in a two-tailed test) means the sample is **unlikely under the null hypothesis**.
- It helps us determine the **p-value** and whether to **reject the null hypothesis**.

---

### üî¢ Example Interpretation:

- **Z = 0** ‚Üí Sample mean exactly equals population mean
- **Z = 2** ‚Üí Sample mean is 2 standard errors above the population mean
- **Z = -2** ‚Üí Sample mean is 2 standard errors below the population mean

---

### üö® Decision Rule (Two-Tailed Test):
- If |Z| > 1.96 ‚Üí **Reject H‚ÇÄ** at 5% significance level
- Otherwise ‚Üí **Fail to reject H‚ÇÄ**

##12. What is the T-test, and How is it Used in Hypothesis Testing?

### üìå What is a T-test?

A **T-test** is a statistical method used to determine if there is a **significant difference between the means** of one or two groups.

It is used when:
- The sample size is **small (n < 30)**
- The **population standard deviation is unknown**
- The data is approximately **normally distributed**

### üéØ Types of T-tests:

|--------------------|--------------------------------------------------------|

### üßÆ Formula (One-sample T-test):

\[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
\]

Where:
- \( \bar{x} \): Sample mean

- \( s \): Sample standard deviation  
- \( n \): Sample size

### ‚úÖ Steps to Use a T-test in Hypothesis Testing:

2. Choose significance level (Œ±), commonly 0.05
3. Calculate the T-statistic
4. Determine the p-value from the T-distribution
5. Compare the p-value with Œ±:
   - If **p-value < Œ±** ‚Üí Reject H‚ÇÄ
   - If **p-value ‚â• Œ±** ‚Üí Fail to reject H‚ÇÄ

---

### üìå Example Use:
Test if the average test score of a sample is different from the expected population mean.

##14. What is a Confidence Interval, and How is it Used to Interpret Statistical Results?

### üìå What is a Confidence Interval?

A **confidence interval (CI)** is a range of values, derived from a sample, that is likely to contain the **true population parameter** (like the mean) with a certain level of confidence.

---

### üßÆ Confidence Interval Formula (for mean):

If the population standard deviation is **unknown**, we use the **T-distribution**:

$$
\text{CI} = \bar{x} \pm t_{\alpha/2} \cdot \left(\frac{s}{\sqrt{n}}\right)
$$

Where:
- \( \bar{x} \): Sample mean  
- \( s \): Sample standard deviation  
- \( n \): Sample size  
- \( t_{\alpha/2} \): t-critical value from t-distribution (based on confidence level and degrees of freedom)

---

### üéØ Interpreting a 95% Confidence Interval:

> "We are 95% confident that the **true population mean** lies within this interval."

This **does not** mean there's a 95% chance that the population mean lies in the interval. Instead, it means **95% of such intervals** would contain the true mean if we repeated the sampling many times.

---

### ‚úÖ Example:

If the 95% confidence interval for average test scores is:

$$
[70.2, 73.8]
$$

We say:
> "We are 95% confident that the average score of the population is between 70.2 and 73.8."

##15.  What is the margin of error, and how does it affect the confidence interval ?

### üìå What is the Margin of Error?

The **margin of error (MoE)** is the amount added and subtracted from the sample statistic (e.g., mean or proportion) to create a **confidence interval**. It represents the **maximum expected difference** between the sample estimate and the true population parameter.

---

### üßÆ Formula (for mean, unknown œÉ):

\[
\text{Margin of Error (MoE)} = t_{\alpha/2} \cdot \left(\frac{s}{\sqrt{n}}\right)
\]

Where:
- \( t_{\alpha/2} \): t-critical value from t-distribution  
- \( s \): Sample standard deviation  
- \( n \): Sample size

---

### üéØ How It Affects the Confidence Interval:

\[
\text{Confidence Interval} = \bar{x} \pm \text{Margin of Error}
\]

- **Larger MoE ‚Üí wider confidence interval** ‚Üí more uncertainty  
- **Smaller MoE ‚Üí narrower confidence interval** ‚Üí more precision  

---

### üìå Factors that Increase Margin of Error:
- Smaller sample size  
- Higher confidence level (e.g., 99% vs 95%)  
- More variability in the data (higher standard deviation)

---

### ‚úÖ Summary:

- The **margin of error controls the width** of the confidence interval.
- It tells us how much the sample estimate might differ from the actual population value.

##16.  How is Bayes' Theorem used in statistics, and what is its significance ?

### üìå What is Bayes' Theorem?

**Bayes' Theorem** is a mathematical formula used to update the **probability of a hypothesis** based on new evidence.

---

### üßÆ Bayes‚Äô Theorem Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Where:
- \( P(A|B) \): Posterior probability (probability of A given B)  
- \( P(B|A) \): Likelihood (probability of B given A)  
- \( P(A) \): Prior probability of A  
- \( P(B) \): Total probability of B

---

### üéØ How Is It Used?

Bayes' Theorem allows us to:
- **Update probabilities** when new information is available.
- Make **informed decisions** in fields like medical diagnosis, spam filtering, and machine learning.
- Incorporate **prior knowledge** into statistical inference.

---

### üìå Example Use Cases:
- **Medical Testing**: Update the probability that a patient has a disease given a positive test result.
- **Email Filtering**: Update the probability that an email is spam based on keywords.
- **Machine Learning**: Used in **Naive Bayes classifiers**.

---

### ‚úÖ Significance in Statistics:

- Helps move from **prior beliefs** to **posterior probabilities**.
- Core of **Bayesian inference**, an alternative to classical frequentist methods.
- Useful when data is limited but prior knowledge is available.

##17.What is the Chi-square Distribution, and When Is It Used?

### üìå What is the Chi-square (œá¬≤) Distribution?

The **Chi-square distribution** is a **skewed, non-negative** distribution that arises when summing the squares of independent standard normal variables.

It is widely used in **inferential statistics**, especially for **categorical data**.

---

### üìä Key Features:

- The shape depends on **degrees of freedom (df)**.
- Always **non-negative** (œá¬≤ ‚â• 0).
- Becomes more symmetric as df increases.

---

### üß™ Common Uses in Statistics:

| Use Case                          | Description                                              |
|----------------------------------|----------------------------------------------------------|
| **Goodness-of-fit test**         | Tests if observed frequencies match expected frequencies |
| **Test for independence**        | Tests if two categorical variables are related (contingency table) |
| **Variance test**                | Tests if sample variance matches a hypothesized value    |

---

### üìå Chi-square Formula (for goodness-of-fit):

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:
- \( O_i \): Observed frequency  
- \( E_i \): Expected frequency  

---

### ‚úÖ Example Applications:
- Do survey results follow a known distribution?
- Are gender and product preference independent?
- Is the variation in a process statistically significant?

---

### üìå Important:
Chi-square tests assume:
- Observations are **independent**
- Expected frequencies are **sufficiently large** (typically ‚â• 5)

---
##18. ## üìä What is the Chi-square Goodness of Fit Test, and How Is It Applied?

### üìå What is the Chi-square Goodness of Fit Test?

The **Chi-square goodness of fit test** is used to determine whether the **observed frequency distribution** of a categorical variable **matches an expected distribution**.

---

### üßÆ Test Statistic Formula:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:
- \( O_i \): Observed frequency for category *i*  
- \( E_i \): Expected frequency for category *i*

---

### ‚úÖ When to Use It:

- You have **one categorical variable**
- You want to test if the observed counts **fit a known distribution** (e.g., uniform, historical data, theoretical ratios)

---

### üß™ Steps to Apply:

1. State hypotheses:
   - H‚ÇÄ: The observed distribution fits the expected distribution
   - H‚ÇÅ: The observed distribution does not fit

2. Calculate expected frequencies

3. Compute the Chi-square test statistic

4. Find the p-value and compare it to Œ± (e.g., 0.05)

5. Make a decision:
   - If p-value < Œ± ‚Üí Reject H‚ÇÄ
   - If p-value ‚â• Œ± ‚Üí Fail to reject H‚ÇÄ

---

### ‚ö†Ô∏è Assumptions:
- Categories are **mutually exclusive**
- **Expected frequency ‚â• 5** in each category (for validity)

##19. What is the F-distribution, and When Is It Used in Hypothesis Testing?

### üìå What is the F-distribution?

The **F-distribution** is a right-skewed probability distribution used primarily to compare **two variances** or test **overall model significance** in regression and ANOVA.

---

### üî¢ Characteristics:

- Always **positive** (F ‚â• 0)
- **Shape depends** on two parameters: degrees of freedom in numerator (df‚ÇÅ) and denominator (df‚ÇÇ)
- Becomes more symmetric as degrees of freedom increase

---

### üß™ Common Uses of the F-distribution:

| Test/Use Case                  | Description                                                                 |
|-------------------------------|-----------------------------------------------------------------------------|
| **ANOVA (Analysis of Variance)** | Tests whether three or more group means are significantly different        |
| **Test for Equality of Variances** | Compares variances of two populations                                       |
| **Multiple Regression (overall F-test)** | Tests whether the regression model as a whole is significant            |

---

### üßÆ F-statistic Formula (ANOVA context):

\[
F = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}} = \frac{MSB}{MSW}
\]

Where:
- MSB = Mean Square Between groups  
- MSW = Mean Square Within groups

---

### ‚úÖ Decision Rule:

- **H‚ÇÄ (null)**: No difference in group means (or equal variances)
- **H‚ÇÅ (alt.)**: At least one mean is different (or variances differ)
- If **F-statistic > critical value** or **p-value < Œ±**, reject H‚ÇÄ

---

### ‚ö†Ô∏è Assumptions:

- Data are normally distributed
- Samples are independent
- Variances are equal across groups (for ANOVA)

##20.  What is an ANOVA test, and what are its assumptions ?

### üìå What is ANOVA?

**ANOVA** (Analysis of Variance) is a statistical test used to determine whether there are **statistically significant differences between the means** of **three or more independent groups**.

It uses the **F-distribution** to assess if the **between-group variance** is significantly larger than the **within-group variance**.

---

### üßÆ Hypotheses:

- **H‚ÇÄ (Null)**: All group means are equal (no difference)
- **H‚ÇÅ (Alternative)**: At least one group mean is different

---

### ‚úÖ ANOVA Test Formula:

$$
F = \frac{MSB}{MSW}
$$

Where:
- **MSB**: Mean Square Between (variance between group means)  
- **MSW**: Mean Square Within (variance within groups)

---

### üß† Assumptions of ANOVA:

1. **Independence** of observations
2. **Normality**: Data in each group should be approximately normally distributed
3. **Homogeneity of variances**: Groups should have similar variances (use Levene‚Äôs Test to check)

---

### üìå Use Cases:

- Compare test scores across 3 teaching methods
- Analyze effectiveness of multiple diets
- Compare product satisfaction across regions

---

### üîç Interpretation:

- If **p-value < Œ±** (e.g., 0.05), reject H‚ÇÄ ‚Üí At least one group mean is different
- If **p-value ‚â• Œ±**, fail to reject H‚ÇÄ ‚Üí No significant difference between means

---
##21. What are the different types of ANOVA tests ?


**ANOVA** (Analysis of Variance) is a family of statistical tests used to determine whether there are **significant differences between group means**. There are **different types of ANOVA** depending on the number of factors and the design of the experiment.

---

### üîπ 1. **One-Way ANOVA**

- üìå Compares the means of **3 or more independent groups** based on **1 independent variable (factor)**
- ‚úÖ Example: Compare average exam scores across 3 different teaching methods.

---

### üîπ 2. **Two-Way ANOVA (Factorial ANOVA)**

- üìå Tests the effects of **two independent variables (factors)** on one dependent variable.
- Also examines if there is an **interaction** between the two factors.
- ‚úÖ Example: Study the effects of **teaching method** and **gender** on test scores.

---

### üîπ 3. **Repeated Measures ANOVA**

- üìå Used when the **same subjects** are measured multiple times under different conditions or time points.
- ‚úÖ Example: Test scores measured **before, during, and after** training on the same group.

---

### üîπ 4. **Mixed-Design ANOVA (Split-Plot ANOVA)**

- üìå Combines **between-subject** (e.g., different groups) and **within-subject** (e.g., repeated measures) factors.
- ‚úÖ Example: Compare different treatment groups over multiple time points.

---

### üßæ Summary Table:

| ANOVA Type               | Factors          | Groups Repeated? | Use Case                                    |
|--------------------------|------------------|------------------|----------------------------------------------|
| **One-Way ANOVA**        | 1                | No               | Compare means of ‚â•3 independent groups       |
| **Two-Way ANOVA**        | 2                | No               | Assess interaction between 2 independent vars|
| **Repeated Measures**    | 1 (or more)      | Yes              | Track same group over time                   |
| **Mixed-Design ANOVA**   | 2 (mixed)        | Some Yes/No      | Compare groups across repeated measures      |

---

### üìå Why So Many?

Each type is designed to **fit different experimental setups** and answer different kinds of questions about the relationships between variables.

##22. What is the F-test, and How Does It Relate to Hypothesis Testing?

### üìå What is the F-test?

The **F-test** is a statistical test used to compare **two variances** or to evaluate whether a group of variables **explains a significant amount of variation** in a model.

It is based on the **F-distribution**, which arises when comparing **ratios of variances**.

---

### üéØ Common Uses of the F-test:

| Type of F-test             | Purpose                                                       |
|----------------------------|---------------------------------------------------------------|
| **ANOVA F-test**           | Compares variances across **three or more group means**       |
| **Variance Ratio Test**    | Compares **two population variances**                         |
| **Regression F-test**      | Tests if the **overall regression model** is significant      |

---

### üßÆ F-statistic Formula (general form):

$$
F = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}} = \frac{MSB}{MSW}
$$

Or for comparing two variances:

$$
F = \frac{s_1^2}{s_2^2}
$$

Where \( s_1^2 \) and \( s_2^2 \) are sample variances from two independent groups.

---

### ‚úÖ Hypothesis Testing with F-test:

- **H‚ÇÄ (Null Hypothesis)**:
  - ANOVA: All group means are equal
  - Variance test: Both groups have equal variances
  - Regression: All regression coefficients = 0 (no effect)
  
- **H‚ÇÅ (Alternative Hypothesis)**: At least one group differs (ANOVA), variances are not equal, or model is significant

- Reject H‚ÇÄ if:
  - \( F \) is **greater than the critical value** from F-distribution
  - or **p-value < significance level (Œ±)** (e.g., 0.05)

---

### üîç Summary:

The **F-test is central to ANOVA and regression analysis**, making it a powerful tool in hypothesis testing when comparing multiple groups or evaluating overall model fit.

##3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1)
        Outcome (0 = failure, 1 = success)
    p : float (0 ‚â§ p ‚â§ 1)
        Probability of success

    Returns:
    float : Probability of x under Bernoulli(p)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")
    return p if x == 1 else (1 - p)

# Example usage:
print("P(X = 1) when p = 0.7:", bernoulli_pdf(1, 0.7))
print("P(X = 0) when p = 0.7:", bernoulli_pdf(0, 0.7))
```

##15.  Implement hypothesis testing using Z-statistics for a sample dataset.


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([68, 70, 72, 71, 69, 73, 70, 68, 72, 74])
sample_mean = np.mean(sample)
n = len(sample)

# Known population parameters
mu_0 = 70     # population mean (hypothesized)
sigma = 3     # known population standard deviation
alpha = 0.05  # significance level

# Step 1: Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 2: Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Decision
if p_value < alpha:
    print("üî¥ Reject the Null Hypothesis (H‚ÇÄ): Significant difference.")
else:
    print("üü¢ Fail to Reject the Null Hypothesis (H‚ÇÄ): No significant difference.")
```

##16.  Create a confidence interval for a dataset using Python and interpret the result.


```python
import numpy as np
from scipy import stats

# Sample data
data = np.array([68, 70, 72, 71, 69, 73, 70, 68, 72, 74])
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level

# Get the t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Calculate margin of error
margin_of_error = t_critical * (std_dev / np.sqrt(n))

# Calculate confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Display results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("We are 95% confident that the true population mean lies within this interval.")
```

##5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=10, n=50, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)
    margin = z_critical * se

    # Rejection regions under H0
    lower_crit = mu_null - margin
    upper_crit = mu_null + margin

    # x-values for plotting
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    # PDFs
    h0_pdf = norm.pdf(x, loc=mu_null, scale=se)
    h1_pdf = norm.pdf(x, loc=mu_alt, scale=se)

    # Plot distributions
    plt.figure(figsize=(12, 6))
    plt.plot(x, h0_pdf, label="H‚ÇÄ: Œº = {}".format(mu_null), color="blue")
    plt.plot(x, h1_pdf, label="H‚ÇÅ: Œº = {}".format(mu_alt), color="green")

    # Fill Type I error region (Œ±) ‚Äî area under H‚ÇÄ in rejection zone
    plt.fill_between(x, 0, h0_pdf, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3, label="Type I Error (Œ±)")

    # Fill Type II error region (Œ≤) ‚Äî area under H‚ÇÅ in acceptance zone
    plt.fill_between(x, 0, h1_pdf, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.3, label="Type II Error (Œ≤)")

    # Critical lines
    plt.axvline(lower_crit, color='red', linestyle='--')
    plt.axvline(upper_crit, color='red', linestyle='--')

    # Labels and legend
    plt.title("üìä Visualization of Type I (Œ±) and Type II (Œ≤) Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Run the function
visualize_type1_type2_errors()
```

##9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin = t_crit * se

    # CI range
    lower = sample_mean - margin
    upper = sample_mean + margin

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print("\nüß† Interpretation:")
    print(f"We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data)
```

##11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


```python
def bayes_theorem(prior, sensitivity, specificity):
    """
    prior: P(Disease) ‚Äî prior probability of the disease
    sensitivity: P(Positive|Disease) ‚Äî test sensitivity
    specificity: P(Negative|No Disease) ‚Äî test specificity
    """
    # Likelihoods
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity

    # Marginal probability of positive test
    p_pos = (p_pos_given_disease * prior) + (p_pos_given_no_disease * (1 - prior))

    # Posterior probability
    posterior = (p_pos_given_disease * prior) / p_pos

    # Results
    print(f"Prior (P(Disease))           : {prior:.4f}")
    print(f"Likelihood (P(+|Disease))     : {sensitivity:.2f}")
    print(f"False Positive Rate (P(+|No)) : {p_pos_given_no_disease:.2f}")
    print(f"Marginal P(Positive)          : {p_pos:.4f}")
    print(f"\nüß† Posterior P(Disease|Positive): {posterior:.4f}")

# Example: Disease affects 1% of population
bayes_theorem(prior=0.01, sensitivity=0.99, specificity=0.95)
```

##12.  Perform a Chi-square test for independence between two categorical variables in Python.


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create contingency table
# Example: Survey of Gender vs Preferred Product
data = pd.DataFrame({
    'Product_A': [30, 20],  # [Male, Female]
    'Product_B': [10, 40]
}, index=['Male', 'Female'])

print("Contingency Table:")
print(data)

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Results
print(f"\nChi-square statistic: {chi2:.3f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

# Step 4: Decision
alpha = 0.05
if p < alpha:
    print("üî¥ Reject the Null Hypothesis: Variables are dependent (associated).")
else:
    print("üü¢ Fail to Reject the Null Hypothesis: No significant association found.")

# Show expected frequencies
print("\nExpected Frequencies Table:")
print(pd.DataFrame(expected, index=data.index, columns=data.columns))
```
